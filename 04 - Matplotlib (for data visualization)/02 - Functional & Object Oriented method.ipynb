# MatplotLib Commands :
Two ways of creating matplotlib plots:

1. Functional Method.
2. Object Oriented Method.

# Functional method

**`plt.plot(x_coordinate, y_coordinate ...n, 'color & line styles attributes')`**

```python
plt.plot(x,y)
# plt.show()    
# you don't have to use unless you're using something that's not jupyter notebook
```

```python
plt.plot(x,y)
plt.show()  
# In this, you don't get the 'Out[]' as in above one, that's the only difference i.e.
# The difference can be related from the analogy 'string' or print('string')
# This is printing out this plot rather than just showing the plot
```

```python
plt.plot(x,y,'r-')    # Adding colour
plt.show()  
```

```python
plt.plot(x,y)

# Labelling axis in using .xlabel(' ') & .ylabel(' ')
plt.xlabel('X Label')
plt.ylabel('Y Label')

# Giving Title to plot using .title(' ')
plt.title('Title')
```

## Creating multiple plots on the same canvas

**`plt.subplot( no_of_Rows, no_of_Cols, Plot_Number )`**

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(x,y,'b*-')
```

# Object Oriented method
### (Much better method)

- In this we will instantiate figure objects and then call methods or attributes from that object.

- The main idea in using the more formal Object Oriented Method is to create figure objects and then just call methods or attributes of that object.

- This approach is better while dealing with a canvas having multiple plots on it.

`figureName = plt.figure()`

`axesName = figureName.add_axes([left, bottom, width, height])`

`axesName.plot(*args, attributes)`

```python
fig = plt.figure()    # An imaginary blank canvas
```

### add_axes(rect)
rect: This parameter is the dimensions [xmin, ymin, dx, dy] of the new axes. It takes the below elements as arguments in the list:

- xmin: Horizontal coordinate of the lower left corner.
- ymin: Vertical coordinate of the lower left corner.
- dx: Width of the subplot.
- dy: Height of the subplot.

```python
# Creating a Figure (EMPTY CANVAS)
fig = plt.figure()

# adding set of axes to figure 
axes = fig.add_axes([0.1,0.1,0.8,0.8])    # [left of axis, bottom of axis, width of axis, height of axis]

# plot on that set of axes
axes.plot(x,y)

# set labels and title in Object Oriented method
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

# Creating Multiple Figures in same canvas using OO:

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
